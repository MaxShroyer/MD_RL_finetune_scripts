# Build Chess QA Datasets (Merged V2)

```python
from pathlib import Path
import os
import subprocess
import sys

def resolve_chess_qa_dir(start: Path) -> Path:
    start = start.resolve()
    for candidate in [start, *start.parents]:
        direct = candidate / 'synth-chess-dataset' / 'build_chess_qa_dataset.py'
        if direct.exists():
            return candidate
        nested = candidate / 'chess_QA' / 'synth-chess-dataset' / 'build_chess_qa_dataset.py'
        if nested.exists():
            return candidate / 'chess_QA'
    raise RuntimeError('Could not locate chess_QA directory')

cwd = Path.cwd()
chess_qa_dir = resolve_chess_qa_dir(cwd)
builder_path = chess_qa_dir / 'synth-chess-dataset' / 'build_chess_qa_dataset.py'
print('cwd:', cwd)
print('chess_qa_dir:', chess_qa_dir)
print('builder:', builder_path)
```

```python
seed = 42
dataset2_dir = chess_qa_dir / 'synth-chess-dataset' / 'rawdatasets' / 'Chess Pieces.v23-raw.coco'
osf_dir = chess_qa_dir / 'synth-chess-dataset' / 'rawdatasets' / 'osfstorage-archive'
output_dir = chess_qa_dir / 'synth-chess-dataset' / 'outputs'
copy_images = True
build_legacy_v1 = False
build_source_specific_v2 = False
mixed_task_only = True
export_hf_dataset = True
push_to_hub = True
hf_repo_id = 'maxs-m87/chess-qa-synth-v1'
hf_private = False
hf_token = os.environ.get('HF_TOKEN', '')
```

```python
cmd = [
    sys.executable,
    str(builder_path),
    '--seed', str(seed),
    '--dataset2-dir', str(dataset2_dir),
    '--osf-dir', str(osf_dir),
    '--output-dir', str(output_dir),
    '--copy-images', str(copy_images).lower(),
    '--build-legacy-v1', str(build_legacy_v1).lower(),
    '--build-source-specific-v2', str(build_source_specific_v2).lower(),
    '--mixed-task-only', str(mixed_task_only).lower(),
    '--export-hf-dataset', str(export_hf_dataset).lower(),
    '--push-to-hub', str(push_to_hub).lower(),
    '--hf-repo-id', hf_repo_id,
    '--hf-private', str(hf_private).lower(),
]
if hf_token:
    cmd.extend(['--hf-token', hf_token])
print('Running command:')
print(' '.join(cmd))
subprocess.run(cmd, check=True, cwd=str(chess_qa_dir))
```

```python
import json

for name in [ 'mixed_tasks_v2_dataset2']: #'piece_position_v2_dataset2',
    metadata = json.loads((output_dir / name / 'metadata.json').read_text())
    stats = json.loads((output_dir / name / 'stats.json').read_text())
    print('\n', name)
    print(' total_rows:', metadata['total_rows'])
    print(' split_counts:', metadata['split_counts'])
    print(' source_inputs:', metadata.get('source_inputs', []))
    print(' source_counts:', metadata.get('source_counts', {}))
    for split in ['train', 'val', 'test']:
        split_stats = stats['split_stats'][split]
        print(' ', split, split_stats['rows'], split_stats['source_counts'])
```
